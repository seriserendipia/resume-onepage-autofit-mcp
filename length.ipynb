```python
with open('D:\PythonEx\casualProject\myresumebuilder\myexperience.md', 'r', encoding='utf-8') as f:
    markdown_content = f.read()
```

```python
import markdown
from weasyprint import HTML, CSS
import os
import sys



# 定义输出文件名
output_pdf_path = "output_auto_fit.pdf"

# 初始字体大小
font_size = 14  # 初始字体大小，单位是pt
# 字体递减步长
font_step = 0.5  # 每次尝试缩小0.5pt

# 循环尝试，直到内容正好一页
while True:
    print(f"尝试用 {font_size}pt 的字体...")

    # 构建 CSS 样式，动态设置字体大小
    style = CSS(string=f'''
        @page {{ size: A4; margin: 1cm; }}
        body {{
            font-family: sans-serif;
            font-size: {font_size}pt;
            line-height: 1.4;
        }}
    ''')

    # 将 Markdown 转换为 HTML
    html_content = markdown.markdown(markdown_content)

    # 用 WeasyPrint 渲染成 PDF
    doc = HTML(string=html_content).render(stylesheets=[style])

    # 获取页数
    page_count = len(doc.pages)

    if page_count == 1:
        # 如果正好一页，就生成文件并退出循环
        doc.write_pdf(output_pdf_path)
        print(f"\n成功！内容完美地排版在了一页中。最终字体大小为 {font_size}pt。")

        # --- 新增的代码块 ---
        # 自动打开生成的文件
        if sys.platform == 'win32':
            os.startfile(output_pdf_path)
        elif sys.platform == 'darwin':
            os.system(f'open {output_pdf_path}')
        else: # Linux
            os.system(f'xdg-open {output_pdf_path}')
        # ---------------------

        break
    elif page_count > 1:
        # 如果超过一页，缩小字体并继续
        font_size -= font_step
    else:
        print("内容太少，无法填充一页。")
        break
    
    # 添加一个防止无限循环的保护，防止字体变得太小
    if font_size <= 8:
        print("\n警告：字体太小，无法继续缩小。请减少文字内容。")
        # break
```

```python
import ipywidgets as widgets
import markdown
import webbrowser
import os

# 需求说明：
# 1. 通过滑杆调整字体大小、标题比例、行间距、页边距，生成 HTML 文件并自动打开浏览器预览。
# 2. 每次参数变化时，自动覆盖并刷新本地 html 文件。

body_font_slider = widgets.FloatSlider(value=14, min=8, max=32, step=0.5, description='正文字体(pt)')
title_ratio_slider = widgets.FloatSlider(value=1.5, min=1.0, max=2.5, step=0.05, description='标题/正文比例')
line_height_slider = widgets.FloatSlider(value=1.4, min=1.0, max=2.0, step=0.05, description='行间距比例')
margin_ratio_slider = widgets.FloatSlider(value=1.0, min=0.5, max=3.0, step=0.05, description='页边距比例')

def render_to_html(body_font, title_ratio, line_height, margin_ratio):
    style = f"""
    <style>
        body {{
            font-family: sans-serif;
            font-size: {body_font}pt;
            line-height: {line_height};
            margin: {margin_ratio}em;
        }}
        h1, h2, h3, h4, h5, h6 {{
            font-size: {body_font * title_ratio}pt;
        }}
    </style>
    """
    html = markdown.markdown(markdown_content)
    full_html = f"<html><head>{style}</head><body>{html}</body></html>"
    html_path = os.path.join(os.getcwd(), 'preview_resume.html')
    with open(html_path, 'w', encoding='utf-8') as f:
        f.write(full_html)
    # 自动打开浏览器预览
    webbrowser.open('file://' + html_path)

ui = widgets.VBox([body_font_slider, title_ratio_slider, line_height_slider, margin_ratio_slider])
out = widgets.interactive_output(
    render_to_html,
    {
        'body_font': body_font_slider,
        'title_ratio': title_ratio_slider,
        'line_height': line_height_slider,
        'margin_ratio': margin_ratio_slider
    }
)

display(ui, out)
```
